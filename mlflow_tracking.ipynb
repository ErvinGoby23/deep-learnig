```python
import mlflow
import mlflow.pytorch
import torch
import numpy as np
```

```python
# Résultats de tous les modèles
experiments = [
    {"name": "CNN_Scratch",    "auc": 0.6285, "f1": 0.0000, "epochs": 3, "lr": 1e-3, "batch_size": 32},
    {"name": "ResNet18",       "auc": 0.6584, "f1": 0.0172, "epochs": 3, "lr": 1e-3, "batch_size": 32},
    {"name": "ViT",            "auc": 0.5854, "f1": 0.0000, "epochs": 3, "lr": 1e-3, "batch_size": 32},
    {"name": "Image_seule",    "auc": 0.6281, "f1": 0.0000, "epochs": 3, "lr": 1e-3, "batch_size": 32},
    {"name": "Texte_seul",     "auc": 0.4879, "f1": 0.0000, "epochs": 3, "lr": 1e-3, "batch_size": 32},
    {"name": "Multimodal",     "auc": 0.6058, "f1": 0.0000, "epochs": 3, "lr": 1e-3, "batch_size": 32},
]

for exp in experiments:
    with mlflow.start_run(run_name=exp["name"]):
        # Hyperparamètres
        mlflow.log_param("model_name", exp["name"])
        mlflow.log_param("epochs", exp["epochs"])
        mlflow.log_param("learning_rate", exp["lr"])
        mlflow.log_param("batch_size", exp["batch_size"])
        mlflow.log_param("image_size", 64)
        mlflow.log_param("train_size", 10000)
        
        # Métriques
        mlflow.log_metric("auc_mean", exp["auc"])
        mlflow.log_metric("f1_macro", exp["f1"])
        
        print(f"Run loggé : {exp['name']} | AUC: {exp['auc']:.4f}")

print("\nTous les runs MLflow enregistrés ")
```
